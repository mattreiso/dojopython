### donuts

Dado um contador inteiro do numero de donuts, retorne uma string com o formato 'Number of donuts: <count>' onde <count> é o numero recebido. 

Entretanto, se o contador for 10 ou mais, use a palavra 'many' ao invés do contador.

    Exemplo: 
    donuts(5) retorna 'Number of donuts: 5'
    donuts(23) retorna 'Number of donuts: many'


```python
def donuts(num):
  if type(num) == int:
    if num >= 10:
      num = 'many'
    print(f'Number of donuts: {num}')
  else:
    print('Colabora, irmão!')


donuts(40)
```

### both_ends

Dada uma string s, retorne uma string feita com os dois primeiros e os dois ultimos caracteres da string original.

    Exemplo: 
    'spring' retorna 'spng'. Entretanto, se o tamanho da string for menor que 2, retorne uma string vazia.

```python
def both_ends(palavra):
  if len(palavra) > 2:
    return palavra[:2] + palavra[-2:]
  else:
    return ''

print(both_ends('melancia'))
print(both_ends('au'))
print(both_ends('céu'))
```

### not_bad

Dada uma string, encontre a primeira aparição das substrings 'not bad', substitua por 'good' e retorne a string resultante.

    Exemplo: 
    'The dinner is not bad!' retorna 'The dinner is good!'

```python
def not_bad(word):
  return word.replace('not bad', 'good')

not_bad('The dinner is not bad! The weather is not bad!')
```

### Dicts

Encontre palavras chaves em um dicionário e apresente o valor de cada palavra. Em seguida, utilizando um frase, faça o somatório de todos os valores das palavras chaves e apresente o resultado

```python
dic_keywords = {
    'amor' : 1,
    'feliz': 1,
    'banana': 0,
    'ódio': -1,
    'raiva': -1
}
```

```python
dic_keywords.get('café', 0)
```

```python
frase = 'Estou muito feliz pois meu amor me deu uma banana mas estava estragada e eu fiquei com raiva'
```

```python
def analisa_sentimento(frase):
  soma = 0
  for palavra in frase.split():
    soma += dic_keywords.get(palavra, 0)

  if soma > 0:
    print('Positivo')
  elif soma == 0:
    print('Neutro')
  else:
    print('Negativo')
```

```python
analisa_sentimento('O Flamengo, ganhou! Estou feliz')
```

## Utilizando dicionários Léxicos

```python
import requests
```

```python
response = requests.get('https://raw.githubusercontent.com/sillasgonzaga/lexiconPT/master/data-raw/SentiLex-lem-PT02.txt').text
```

```python
print(response)
```

```python
dic_polaridade = {}
for linha in response.split('\n'):
  palavra = linha.split('.')[0]
  polaridade = linha.split(';')[2].split('=')[1]
  dic_polaridade[palavra] = int(polaridade)
```

```python
dic_polaridade
```

```python
def analisa_sentimento(frase):
  soma = 0
  for palavra in frase.split():
    soma += dic_polaridade.get(palavra, 0)

  if soma > 0:
    print('Positivo')
  elif soma == 0:
    print('Neutro')
  else:
    print('Negativo')
```

```python
analisa_sentimento('Estou preucupado com as provas')
```
